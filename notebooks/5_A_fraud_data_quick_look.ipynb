# Quick look at fraud data
# Objective: Gain insight on the new released fraud data

```python
# importing require libraries to run the code in this notebook
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
```

```python
# running spark
spark = (
    SparkSession.builder.appName("preprocessing of taxi data")
    .config("spark.sql.repl.eagerEval.enabled", True) 
    .config("spark.sql.parquet.cacheMetadata", "true")
    .config("spark.sql.session.timeZone", "Etc/UTC")
    .config("spark.driver.memory", "15g")
    .getOrCreate()
)
```

```python
# reading in data
transaction_sdf = spark.read.parquet("../data/tables/transactions_20220228_20220828_snapshot/")
consumer_fraud = spark.read.option("header",True).csv("../data/tables/consumer_fraud_probability.csv")
merchant_fraud = spark.read.option("header",True).csv("../data/tables/merchant_fraud_probability.csv")
```
