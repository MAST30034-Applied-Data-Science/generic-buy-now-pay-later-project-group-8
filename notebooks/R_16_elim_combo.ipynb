```python
import pandas as pd
```

```python
# data = pd.read_csv('../data/curated/final_model/5_vars_MeanStdsdCorrLoyalPersona.csv')
data = pd.read_csv('../data/curated/final_model/5_vars_SharpeCorrGrowthLoyalPersona.csv')
# data = pd.read_csv('../data/curated/final_model/5_vars_StdmeanSdCorrLoyalPersona.csv')
```

```python
data.sort_values(['sharpe_ratio'],ascending=False)
```

```python
data = data[data['sharpe_ratio'] == data.sort_values(['sharpe_ratio'],ascending=False)['sharpe_ratio'][0]]
data
```

```python
combos = list(data['all_combos'])
```

```python
str_combos = [combo.strip('[]').split(',') for combo in combos]
```

```python
float_combos = list()
for combo in str_combos:
    tmp = [float(x.strip(' ')) for x in combo]
    float_combos.append(tmp)
```

```python
remove_ratio = list()
for combo in float_combos:
    tmp = [x/combo[0] for x in combo]
    if combo[0] > 0:
        ratio = 1
    else:
        ratio = -1
    tmp.append(ratio)
    remove_ratio.append(tmp)
```

```python
a1 = [x[0] for x in remove_ratio]
a2 = [x[1] for x in remove_ratio]
a3 = [x[2] for x in remove_ratio]
a4 = [x[3] for x in remove_ratio]
a5 = [x[4] for x in remove_ratio]
ratio = [x[5] for x in remove_ratio]

remove_ratio_df = pd.DataFrame({'Sharpe': a1, 'Hist corr rev': a2, 'growth': a3, 'loyalty':a4, 'persona':a5, 'ratio': ratio})
```

```python
remove_ratio_df.sort_values(by=['Hist corr rev', 'growth', 'loyalty', 'persona', 'ratio']).drop_duplicates()
```

```python
remove_ratio_df.sort_values(by=['Hist corr rev', 'growth', 'loyalty', 'persona']).drop_duplicates().to_csv('../data/tuning/5_vars_SharpeCorrGrowthLoyalPersona_top.csv', index=False)
```
